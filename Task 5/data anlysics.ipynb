```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (8,5)

# Load dataset
df = pd.read_csv("/content/titanic.csv")  # Replace with your file path

```

```python
# Handle missing values for plotting
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# AGE DISTRIBUTION
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
print("""
Observation:
- Most passengers are aged 20–40.
- Very few are below 10 or above 60.
- Slight right skew (fewer older passengers).
""")
```

Observation:
- Most passengers are aged 20–40.
- Very few are below 10 or above 60.
- Slight right skew (fewer older passengers).


```python
# FARE BOXPLOT
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Fare'])
plt.title("Fare Boxplot")
plt.show()
print("""
Observation:
- Fare distribution is right-skewed.
- Several high outliers exist.
- Most passengers paid below ~100.
""")
```

Observation:
- Fare distribution is right-skewed.
- Several high outliers exist.
- Most passengers paid below ~100.


```python
# AGE VS FARE SCATTERPLOT
plt.figure(figsize=(8,6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare")
plt.show()
print("""
Observation:
- Most passengers paid low to mid-range fares.
- High-fare passengers exist across ages.
- Higher fare sometimes correlates with survival.
""")
```

Observation:
- Most passengers paid low to mid-range fares.
- High-fare passengers exist across ages.
- Higher fare sometimes correlates with survival.


```python
# CORRELATION HEATMAP
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
print("""
Observation:
- Fare and Pclass negatively correlated (higher class = higher fare).
- Age has little correlation with Survived.
- Pclass negatively correlates with Survived (lower class = lower survival).
""")
```

Observation:
- Fare and Pclass negatively correlated (higher class = higher fare).
- Age has little correlation with Survived.
- Pclass negatively correlates with Survived (lower class = lower survival).


```python
# PAIRPLOT
sns.pairplot(df[['Age','Fare','Survived']])
plt.show()
print("""
Observation:
- Relationships between Age, Fare, Survived visualized.
- Age and Fare mostly low/mid-range.
- Survived distribution shows no strong pattern with Age alone.
""")
```

Observation:
- Relationships between Age, Fare, Survived visualized.
- Age and Fare mostly low/mid-range.
- Survived distribution shows no strong pattern with Age alone.


Summary of Findings:
The Titanic dataset reveals that the majority of passengers were aged 20–40. Females had a significantly higher survival rate than males. Higher ticket fares were strongly associated with better survival, reflecting that passengers from higher classes were prioritized during rescue. While age alone did not strongly affect survival, passengers who paid higher fares survived more often. Overall, socio-economic status (class and fare) and gender were the most important factors influencing survival.
